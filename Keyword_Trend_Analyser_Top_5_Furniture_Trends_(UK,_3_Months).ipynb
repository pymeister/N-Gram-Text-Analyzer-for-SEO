# Furniture - Rising Trends⚡

Related Queries

````
#@title
kw_list=['sofas', 'beds', 'wardrobes', 'tables', 'furniture']
pytrends.build_payload(kw_list, timeframe='today 3-m', geo='GB')
````

````
#@title
related_queries = pytrends.related_queries()
#related_queries.values()
````

````
#@title
related_queries.get('furniture').get('rising')
````
